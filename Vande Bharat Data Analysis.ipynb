```python
sns.barplot( x=Vande_bharat["Train Name"].value_counts(),
            y=Vande_bharat["Train Name"].value_counts().keys(),
             )
plt.ylabel("Train Name")
plt.xlabel("No. of Trains")
```

```python
sns.barplot( x=Vande_bharat["Originating City"].value_counts(),
            y=Vande_bharat["Originating City"].value_counts().keys(),
             )
plt.ylabel("Originating City")
plt.xlabel("No. of Trains")
```

```python
sns.barplot( x=Vande_bharat["Originating Station"].value_counts(),
            y=Vande_bharat["Originating Station"].value_counts().keys(),
             )
plt.ylabel("Originating Station")
plt.xlabel("No. of Trains")
```

```python
sns.barplot( x=Vande_bharat["Terminal City"].value_counts(),
            y=Vande_bharat["Terminal City"].value_counts().keys(),
             )
plt.ylabel("Terminal City")
plt.xlabel("No. of Trains")
```

```python
sns.barplot( x=Vande_bharat["Operator"].value_counts(),
            y=Vande_bharat["Operator"].value_counts().keys(),
             )
plt.ylabel("Operator")
plt.xlabel("No. of Trains")
```

```python
Vande_bharat["No. of Cars"].value_counts()
```

```python
sns.barplot( x=Vande_bharat["No. of Cars"].value_counts(),
            y=Vande_bharat["No. of Cars"].value_counts().keys(),
             )
plt.ylabel("No. of Trains")
plt.xlabel("No. of Cars")
```

```python
plt.pie( x=Vande_bharat["No. of Cars"].value_counts(), explode=[0.1,0],
       labels=Vande_bharat["No. of Cars"].value_counts().index,
       autopct="%.0f%%",
       colors=["r","b"],
       shadow=True)
plt.title("No. of Cars")
plt.legend()
```

```python
sns.barplot( x=Vande_bharat["Frequency"].value_counts(),
            y=Vande_bharat["Frequency"].value_counts().keys(),
             )
plt.title("Frequency of Train")
plt.ylabel("Frequency")
plt.xlabel("No. of Trains")
```

```python
Vande_bharat["Average occupancy"] = Vande_bharat["Average occupancy"].str.rstrip("%").astype(float)
```

```python
sns.barplot( x=Vande_bharat["Train Name"],
            y=Vande_bharat["Average occupancy"],
             )
plt.title("Average occupancy vs Train")
plt.xticks(rotation=90)
plt.ylabel("Average occupancy")
plt.xlabel("Trains Name")
```

```python
sns.barplot( x=Vande_bharat["Train Name"],
            y=Vande_bharat["No. of Cars"],
             )
plt.title("No. of Cars vs Train")
plt.xticks(rotation=90)
plt.ylabel("No. of Cars")
plt.xlabel("Trains Name")
```
